```python
from pathlib import Path
from mutagen.mp4 import MP4
import pandas as pd
target_dir = Path(r'data\Brainrot')

if target_dir.exists():
    print(True)

df = pd.DataFrame()

for ind, item in enumerate(target_dir.iterdir()):
    mp4 = MP4(item)
    dct = {
    'title': f'Brainrot_{ind}',
    'bitrate': mp4.info.bitrate,
    'length': mp4.info.length,
    'sample_rate': mp4.info.sample_rate,
    'channels': mp4.info.channels
    }

    df_t = pd.DataFrame([dct])
    df = pd.concat([df,df_t],ignore_index=True)
    
```

```python
import os

for ind, file in enumerate(target_dir.iterdir()):

    rep_name = f'data\Brainrot\Brainrot_{ind}.mp4'
    os.rename(file, rep_name, src_dir_fd=None, dst_dir_fd=None)
    print(file)
    
```
