```python


def get_video_info(path:str, output_dir:str):

    vid_cap = cv2.VideoCapture(path)
    vid_num = int(path.split('.')[0].split('_')[-1])

    total_frames = vid_cap.get(cv2.CAP_PROP_FRAME_COUNT)
    indx = np.linspace(0, total_frames, num=16, endpoint=True, dtype=int)
    saved_paths = []

    for i in indx:
        vid_cap.set(cv2.CAP_PROP_POS_FRAMES, i)
        ret, frame = vid_cap.read()
        if not ret:
            continue
        frame = cv2.resize(frame, (224, 224))
        frame_path = os.path.join(output_dir, f"{vid_num:02d}_frame_{i:02d}.jpg")
        cv2.imwrite(frame_path, frame)
        saved_paths.append(frame_path)


    print(f"Saved {len(saved_paths)} frames for video {vid_num}")
    vid_cap.release()
    return saved_paths

output_dir = r'data/frames_brairot/'
b_dir_path = r'data/Brainrot/'

for f in Path(b_dir_path).iterdir():
    get_video_info(str(f), output_dir)
    
```

```python
class CustomDataset(Dataset):
    def __init__(self, data_dir: Path, annotations_file: Path, transform=None):
        self.data_dir = data_dir
        self.annotations_file = annotations_file
        self.transform = transform
        self.video_clips = VideoClips(str(data_dir), clip_length_in_frames=16)

    def __len__(self):
        return len(self.video_clips)

    def __getitem__(self, idx):
        video, _, _, _ = self.video_clips.get_clip(idx)
        return video
```
