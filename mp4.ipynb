#### Making dataframe with metadata of an mp4 file:

Metadata
- bitrate: file bitrate 
- length: length of a record in mls
- sample_rate: sample rate for a record
- channels: number of audio channels

```python
if b_target_dir.exists() and nb_target_dir.exists():
    print('Both directories are found and loaded')

df = pd.DataFrame()

for ind, item in enumerate(itertools.chain(nb_target_dir.iterdir(), b_target_dir.iterdir())):
    mp4 = MP4(item)

    dct = {
    'title': item.name,
    'bitrate': mp4.info.bitrate,
    'length': mp4.info.length,
    'sample_rate': mp4.info.sample_rate,
    'channels': mp4.info.channels
    }

    df_t = pd.DataFrame([dct])
    df = pd.concat([df,df_t],ignore_index=True)
```

```python
df.to_csv('data/metadata.csv', index=False)
```
